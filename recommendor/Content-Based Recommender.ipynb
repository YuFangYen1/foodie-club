```python
import pandas as pd
```

```python
meta_data = pd.read_excel('List_Of_Restaurants.xlsx')
meta_data.head(5)
```

```python
C = meta_data['Rating_out_of_5'].mean()
C
```

```python
m = meta_data['Review_number'].quantile(0.9)
m
```

```python
q_restaurants = meta_data.copy().loc[meta_data['Review_number'] >= m]
q_restaurants.shape
```

```python
meta_data.shape
```

```python
def weighted_rating(x, m=m, C=C):
    v = x['Review_number']
    R = x['Rating_out_of_5']
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
q_restaurants['score'] = q_restaurants.apply(weighted_rating, axis=1)
q_restaurants.head(20)
```

```python
q_restaurants = q_restaurants.sort_values('score', ascending=False)
q_restaurants[['Restaurant_name', 'Rating_out_of_5', 'Review_number', 'score']].head(20)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(stop_words='english')
meta_data['About Section'] = meta_data['About Section'].fillna('')
tfidf_matrix = tfidf.fit_transform(meta_data['About Section'])
tfidf_matrix.shape
```

```python
tfidf.get_feature_names()[1000:1010]
```

```python
from sklearn.metrics.pairwise import linear_kernel
```

```python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
cosine_sim.shape
```

```python
cosine_sim[1]
```

```python
indices = pd.Series(meta_data.index, index=meta_data['Restaurant_name']).drop_duplicates()
indices[:10]
```

```python
def get_recommendations(name, cosine_sim=cosine_sim):
    idx = indices[name]
    
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    sim_scores = sim_scores[1:11]
    
    restaurants_indices = [i[0] for i in sim_scores]
    
    return meta_data['Restaurant_name'].iloc[restaurants_indices]
```

```python
get_recommendations('Kenza')
```
